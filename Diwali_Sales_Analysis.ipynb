```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

```

```python
df.dropna(inplace=True)
```

**📊 Age Group wise distribution**

```python
plt.figure(figsize=(10,5))
ax =sns.countplot(x='Age Group',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
plt.title('Age Group-wise Customer Distribution')
plt.show()
```

**📊 Total Sales by Marital Status and Gender**

```python
sales_demo = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize' :(7,5)})
ax = sns.barplot(data=sales_demo, x='Marital_Status', y='Amount', hue='Gender')
plt.title('Sales by Marital Status & Gender')
plt.tight_layout()
for bars in ax.containers:
  ax.bar_label(bars, fmt='%.0f')

```

**📊 Gender-wise Distribution across Age Group**

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(x='Age Group', hue='Gender', data=df)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Gender-wise Distribution across Age Groups')
plt.tight_layout()
plt.show()
```

**📊 Average Order Value by Age Group and Gender**




```python
avg_order_value = df.groupby('Age Group')['Amount'].mean().sort_values(ascending=False)
ax= avg_order_value.plot(kind='bar', title='Average Order Value by Age Group', figsize=(10, 5))
ax.bar_label(ax.containers[0])
plt.ylabel("Average Order Value")
plt.tight_layout()
plt.show()
```

**📌Top 10 Product Categories by Total Sales**

```python
sales_category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().nlargest(10,'Amount')
plt.figure(figsize=(16, 5))
ax=sns.barplot(data=sales_category,x='Product_Category',y='Amount')
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title('Top 10 Product Categories by Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

****

**📊 Analyzing which age group contributes the most to total sales.**

```python
plt.figure(figsize=(10, 5))
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax= sns.barplot(x='Age Group',y='Amount',data=sales_age)
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title('Total sales by Age Group')
plt.tight_layout()
```

**📌 Top 10 States by Number of Orders**

```python
plt.figure(figsize=(17,6))
order_by_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Orders',data=order_by_state)
ax.bar_label(ax.containers[0])
plt.title('Top 10 States by Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xticks(rotation=45)
```

**📌 Top 10 States by Total Sales**





```python
sales_by_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax = sns.barplot(x='State',y='Amount',data=sales_by_state)
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title('Top 10 States by Sales Revenue')
plt.tight_layout()
plt.show()
```

**📊 Total Sales by Occupation**

```python
sales_by_occupation = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(19, 5))
ax = sns.barplot(x='Occupation', y='Amount', data=sales_by_occupation)
ax.bar_label(ax.containers[0], fmt='%.0f')
plt.title('Total Sales by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 **📌Top 10 Product by Total Sales (Amount)**

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
ax = sns.barplot(data = sales_state, x = 'Product_ID',y= 'Amount')
for bars in ax.containers:
 ax.bar_label(bars)
plt.title('Top 10 Most Sold Products by Order Volume')
```

**📊 Revenue Share by Gender - Pie Chart**

```python
gender_sales = df.groupby('Gender')['Amount'].sum()
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Contribution by Gender')
plt.show()
```

**📈 Heatmap : Occupation vs Product Category**

```python
pivot_table = df.pivot_table(values='Amount', index='Occupation', columns='Product_Category', aggfunc='sum')

plt.figure(figsize=(20, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.0f')

plt.title('Sales Distribution by Occupation and Product Category')
plt.ylabel('Occupation')
plt.xlabel('Product Category')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
segmentation_data = df.groupby(['Marital_Status', 'Gender']).size().unstack(fill_value=0)
segmentation_data.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Set2')

plt.title('Customer Segmentation by Marital Status and Gender')
plt.xlabel('Marital Status')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

##**📌Conclusion and Insights**

*This Diwali Sales analysis reveals strong business patterns that can guide effective marketing and operational strategies:*

*🎯 Target Audience: Females aged 26–45 form the core customer base, with single women contributing the highest sales volume*

*💰 Top Spenders: The 51–55 age group has the highest average order value, suggesting older customers have stronger purchasing power*

**🏆 Best Performing Segments**

*Product Category: Food tops the sales chart*

*Occupation: IT professionals spend the most, especially on Electronics & Clothing.*

*State: Uttar Pradesh leads in both orders and revenue*

**📉 Underperforming Areas:**

*0–17 age group, agriculture occupation, and veterinary category have minimal engagement.*

**📌 Strategic Takeaway: Focus campaigns on working professionals (especially IT & Healthcare), optimize inventory for popular categories, and tailor offers for the high-potential female customer base aged 26–45.**
